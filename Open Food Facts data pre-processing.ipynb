```python
# Install required data
import pandas as pd

import datetime
```

```python
# read in data from download location
data =pd.read_csv('/Users/francescapontin/Downloads/en.openfoodfacts.org.products.tsv', sep='\t')
```

## Clean and sub-set the data

```python
dtypes={'created_datetime':datetime,
'last_modified_datetime':datetime,
'product_name':object,
'generic_name':object,
'quantity':object,
'serving_size':object,
'nutrition_grade_fr':object,
'pnns_groups_1':'category',
'pnns_groups_2':'category',
'main_category_en':object,
'energy_100g':'float64',
'fat_100g':'float64',
'saturated-fat_100g':'float64',
'monounsaturated-fat_100g':'float64',
'polyunsaturated-fat_100g':f'loat64',
'omega-3-fat_100g':'float64',
'omega-6-fat_100g':'float64',
'omega-9-fat_100g':'float64',
'trans-fat_100g':'float64',
'cholesterol_100g':'float64',
'carbohydrates_100g':'float64',
'sugars_100g':'float64',
'starch_100g':'float64',
'fiber_100g':'float64',
'proteins_100g':'float64'}
```

```python
# Only include data with a UK nutri-score
data=data.loc[~data['nutrition-score-uk_100g'].isna(),:]

# Only include data with generic name info
data=data.loc[~data['generic_name'].isna(),:]

# Only include data where quantity is known
data=data.loc[~data['quantity'].isna(),:]

# specify data types
data[['created_datetime','last_modified_datetime']] =data[['created_datetime','last_modified_datetime']].apply(pd.to_datetime)
data['pnns_groups_1']=data['pnns_groups_1'].astype("category")
data['pnns_groups_2']=data['pnns_groups_2'].astype("category")


```

```python
# Subset columns to only include relevant ones for exploratory analysis and to reduce file size
data[['created_datetime','last_modified_datetime','product_name', 'generic_name','quantity','serving_size',
 'nutrition_grade_fr',
 'nutrition-score-fr_100g',
 'pnns_groups_1',
 'pnns_groups_2',
 'main_category_en',
 'energy_100g',
 'fat_100g',
 'saturated-fat_100g',
 'monounsaturated-fat_100g',
 'polyunsaturated-fat_100g',
 'omega-3-fat_100g',
 'omega-6-fat_100g',
 'omega-9-fat_100g',
 'trans-fat_100g',
 'cholesterol_100g',
 'carbohydrates_100g',
 'sugars_100g',
'starch_100g',
 'fiber_100g',
 'proteins_100g']].to_csv('data/Open_Food_facts_sample_data.csv')
#, dtype =dtypes)
```
